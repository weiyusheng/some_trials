```python
from functools import total_ordering

from sortedcontainers import SortedSet

@total_ordering
class ran:
    def __init__(self, l ,r=-1 ,v=0):
        self.l = l
        self.r = r
        self.v = v
    def __lt__(self,o_ran):
        return self.l < o_ran.l
    def __eq__(self,o_ran):
        return self.l == o_ran.l
    def __hash__(self) -> int:
        return hash(self.l)
    def __repr__(self) -> str:
        return f"l={self.l} r={self.r} v={self.v}"
    def __str__(self) -> str:
        return f"l={self.l} r={self.r} v={self.v}"

seed:int = None
def rnd():
    global seed
    ret = seed
    seed = (seed * 7 + 13) % 1000000007
    return ret

class CT(SortedSet):
    def __init__(self, n, vmax):
        super().__init__(self)
        for i in range(n):
            temp = rnd()
            temp = temp % vmax + 1
            self.add(ran(i + 1, i + 1, temp))
        self.add(ran(n + 1, n + 1, 0))

    def step(self, item, forward=True):
        if forward:
            return self[self.index(item)+1]
        else:
            return self[self.index(item)-1]

    def split(self, pos):
        it = self.bisect_left(ran(pos))
        if it != len(self) - 1 and self[it].l == pos:
            return self[it]
        it:ran = self.step(self[it], False)
        l = it.l
        r = it.r
        v = it.v
        self.remove(it)
        self.add(ran(l, pos - 1, v))
        tem = ran(pos, r, v)
        self.add(tem)
        return tem

    def assign(self, l, r, v):
        it2 = self.split(r + 1)
        it1 = self.split(l)
        while it1 != it2:
            tem = it1
            it1 = self.step(it1)
            self.remove(tem)
        self.add(ran(l, r, v))

    def adds(self, l, r, v):
        it2 = self.split(r + 1)
        it1 = self.split(l)
        while it1 != it2:
            it1.v += v
            it1 = self.step(it1)

    def kth(self, l, r, k):
        it2 = self.split(r + 1)
        it1 = self.split(l)
        while it1 != it2:
            k -= (it1.r - it1.l + 1)
            if k <= 0:
                return it1.v
            it1 = self.step(it1)

    def query(self, l, r, x, vmax):
        it2 = self.split(r + 1)
        it1 = self.split(l)
        res = 0
        while it1 != it2:
            res += (it1.r - it1.l + 1) * pow(it1.v, x, vmax)
            it1 = self.step(it1)
        return res

    def __str__(self):
        tem = ''
        for item in self:
            for len in range(item.r - item.l + 1):
                tem += f"{item.v},"
        return tem[:-1]

    def work(self, n, m, vmax):
        for i in range(m):
            op = rnd() % 4 + 1
            l = rnd() % n + 1
            r = rnd() % n + 1
            x = -1
            y = -1
            if l > r:
                tem = l
                l = r
                r = tem
            if op == 3:
                x = rnd() % (r - l + 1) + 1
            else:
                x = rnd() % vmax + 1
            if op == 4:
                y = rnd() % vmax + 1
            if op == 1:
                self.adds(l, r, x)
            elif op == 2:
                self.assign(l, r, x)
            elif op == 3:
                print(self.kth(l, r, x))
            elif op == 4:
                print(self.query(l, r, x, y) % y)

if __name__ == "__main__":
    # inputs = input().split()
    # n = int(inputs[0])
    # m = int(inputs[1])
    # seed = int(inputs[2])
    # vmax = int(inputs[3])

    n = 10
    m = 10
    seed = 7
    vmax = 9

    C_T = CT(n, vmax)
    C_T.work(n, m, vmax)
    # 输入格式是按那道题来的，至于哪道题，上面贴了个链接
    # 没有经过完全的测试，仅供参考
    # 谁写oj用python啊
```
